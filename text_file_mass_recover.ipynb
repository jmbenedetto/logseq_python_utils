# PACKAGES

```python
import os
import re
```

# PARAMETERS

```python
backup_directory = "/Users/joao.benedetto/logseq_graph/logseq/bak/notes"
target_directory = "/Users/joao.benedetto/logseq_graph/notes"
change_time = {"start": 1728775440.0, "end": 1728775560.0}
```

# FUNCTIONS

```python
def move_and_rename_files(backup_dir, target_dir, change_time):
    import os
    import shutil
    from datetime import datetime

    # Iterate through each subdirectory in the backup directory
    for subdir in os.listdir(backup_dir):
        subdir_path = os.path.join(backup_dir, subdir)
        # print(f"Checking subdirectory: {subdir_path}")

        # Check if it's a directory
        if os.path.isdir(subdir_path):
            # print(f"{subdir_path} is a directory")

            # Iterate through each file in the subdirectory
            for file in os.listdir(subdir_path):
                file_path = os.path.join(subdir_path, file)
                # print(f"Checking file: {file_path}")

                # Check if the file was last modified at the specified change_time
                if (
                    os.path.getmtime(file_path) >= change_time["start"]
                    and os.path.getmtime(file_path) <= change_time["end"]
                ):
                    # print(f"File {file_path} was last modified at the specified change_time")

                    # Construct the new file name with .md extension
                    new_file_name = f"{subdir}.md"
                    target_file_path = os.path.join(target_dir, new_file_name)
                    # print(f"New file name will be: {new_file_name}")

                    # Remove the existing file in the target directory if it exists
                    if os.path.exists(target_file_path):
                        os.remove(target_file_path)
                        # print(f"Removed existing file at {target_file_path}")

                    # Move and rename the file to the target directory
                    shutil.move(file_path, target_file_path)
                    print(f"Moved and renamed {file} to {target_file_path}")
                    break


# Example usage
# move_and_rename_files(backup_directory, target_directory, some_change_time)
```

# SCRIPT


```python
move_and_rename_files(backup_directory, target_directory, change_time)
```
